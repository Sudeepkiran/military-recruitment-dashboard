```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("data/military_data.csv")
df.head()
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(data=df, x="Year", y="Recruits", hue="Branch", marker="o")
plt.title("Military Recruits Over the Years by Branch")
plt.ylabel("Number of Recruits")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x="Branch", y="Attrition Rate (%)", ci=None)
plt.title("Average Attrition Rate by Branch")
plt.ylabel("Attrition Rate (%)")
plt.grid(True)
plt.show()
```

```python
army_df = df[df["Branch"] == "Army"]
X = army_df[["Year"]]
y = army_df[["Recruits"]]

model = LinearRegression()
model.fit(X, y)

pred_2025 = model.predict([[2025]])
print(f"Predicted Army Recruits in 2025: {int(pred_2025[0][0])}")
```
